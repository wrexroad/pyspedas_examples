# Inspecting magnetometer data using URL builder

Or you can build URL request check data availability via [https://service.iris.edu/fdsnws/availability/docs/1/builder/](https://service.iris.edu/fdsnws/availability/docs/1/builder/)

For magnetometer data the second letter in the channel code should be **F**. 
First letter will correspond to the band, for example:
M Mid Period > 1 to < 10 Hz
L Long Period ≈ 1 Hz
V Very Long Period ≈ 0.1 Hz
U Ultra Long Period ≈ 0.01 Hz
Third letter correspond to the orientation and should be **Z**, **N**, or **E** (Vertical, North-South, East-West).

Using the wiledcards, the channel can be specifyed as `?FE,?FN,?FZ`.

The location must be specified as `--`, which is the only one that is supported at this time.

Example of URL of magnetometer data availability for 8 October 2012 storm :
`https://service.iris.edu/fdsnws/availability/1/query?format=text&loc=--&cha=?FE,?FN,?FZ&starttime=2012-10-08&endtime=2012-10-09&orderby=nslc_time_quality_samplerate&includerestricted=false&nodata=404`

```python
import urllib
url = 'https://service.iris.edu/fdsnws/availability/1/query?format=text&net=*&sta=*&loc=--&cha=?FE,?FN,?FZ&starttime=2012-10-08&endtime=2012-10-09&orderby=nslc_time_quality_samplerate&includerestricted=false&nodata=404'
with urllib.request.urlopen(url) as response:
   html = response.read().decode('utf-8')
print(html)
```
