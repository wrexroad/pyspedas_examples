```python
from pytplot import store_data, tplot, options
```

Create a simple tplot variable named data with all values set to 1


```python
store_data('data', data={'x': [1, 2, 3, 4, 5, 6], 'y': [1, 1, 1, 1, 1, 1]})
```

By default, no markers are set

```python
tplot('data')
```

Change the marker to `X`

```python
options('data', 'marker', 'X')
```

```python
tplot('data')
```

Change the plot to use symbols to disable the line connecting the markers. Setting 'symbols' to True creates a matplotlib scatter plot, rather than a line plot.

```python
options('data', 'symbols', True)
```

```python
tplot('data')
```

Change the `marker_size` option to increase the size of the symbols

```python
options('data', 'marker_size', 256)
```

```python
tplot('data')
```

Disable `symbols` to enable to line connecting the data points

```python
options('data', 'symbols', False)
tplot('data')
```

Lower the marker size. Note: the definition of `marker_size` seems to be different depending on if the `symbols` option is set. To match the size of scatter (symbols=True) and plot markers, the plot marker's markersize should be the square root of the scatter (symbol=True) marker size. Resetting the marker_size to 16 results in the same size markers as were displayed in the 'symbols' plot with marker_size 256.



```python
options('data', 'marker_size', 16)
```

```python
tplot('data')
```

Another way to disable the connecting lines is to set the linestyle to 'None'.   This is useful when a matplolib scatter plot might not be appropriate, such as a plot displaying error bars.

```python
options('data','line_style','None')
tplot('data')
```

Update the figure to only show the marker every 2 data points

```python
options('data', 'markevery', 2)
```

```python
tplot('data')
```

Change the markers to hexagons.

Note: this supports all of the same markers shown at:

https://matplotlib.org/stable/api/markers_api.html

```python
options('data', 'marker', 'H')
```

```python
tplot('data')
```
